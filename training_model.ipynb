## Creating predictors for machine learning

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches.head()
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches.head()
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["target"] = (matches["result"] == "W").astype("int") ##returns true for win and converts to one
```

```python
# creating the initial machine learning model
```

```python
from sklearn.ensemble import RandomForestClassifier # machine learning model that can pick up non-linear relationships
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split = 10, random_state=1)
```

```python
train = matches[matches["date"] < '2023-01-01']
```

```python
test = matches[matches["date"] > '2023-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"]) # train a model with these predictors trying to fit "target" values
```

```python
preds = rf.predict(test[predictors])
```

```python
# determining the accuracy

from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds) # actuals, predictions
```

```python
# improving predictions with moving averages - eg, how will they do in match week 4 given their 3 previous matches
```

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
def rolling_average(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed = "left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #if you are in week 2, there are not 3 prev weeks
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_average(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(
    lambda x: rolling_average(x.reset_index(), cols, new_cols), include_groups=False
)
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```
